```python
# First try using json.load
import json

with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    data = json.load(raw_file)
    print(data)
```

```python
# It's a collection of documents, json.load only can read a file with 
#  a single document inside it, so, let's try another solution 
#  using json.loads (s of string and the format it to dict)
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    data = json.loads(raw_file)
    print(data)
```

```python
# The error: the JSON object must be str, bytes or bytearray, not TextIOWrapper
#  means that you need to get an str object, if you check the type of our data 
#  variable will by TextIOWrapper
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    print(type(raw_file)) # <class '_io.TextIOWrapper'>
    
```

```python
# So... how we can get an str object from the file ? - Reading it line by line !
#  let's process the data with a foor loop
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    for line in raw_file:
        print(type(line)) # <class 'str'>
```

```python
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    current_line = 1 # Yes, only 100 rows (line) to avoid IOPub data rate exceeded. error
    for line in raw_file:
        print(line)
        
        if current_line < 100:
            current_line += 1
        else:
            break
```

```python
# Ok now we can reach each line, now, let's try to decode each line as a json object
#  using json.loads function
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    for line in raw_file:
        print(json.loads(line)) # Expecting property name enclosed in double quotes: line 1 column 2 (char 1)
```

```python
# Mmmm ok, double quotes, easy ! just replace them in each line
with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    
    for line in raw_file:
        current_line = line
        current_line = current_line.replace("\'", "\"")
        
        print(current_line)
        print(json.loads(current_line)) # Expecting ',' delimiter: line 1 column 318 (char 317)
```

```python
# We can also try using raw text (r"") but, exists any other solution less complex ?
#  Research about ast module with literal_eval function

import ast
```

```python
%%time
raw_qa = []

with open("./sources/qa_Cell_Phones_and_Accessories.json") as raw_file:
    for line in raw_file:
        current_line = ast.literal_eval(line)
        raw_qa.append(current_line)
        
# Finally !!
```
