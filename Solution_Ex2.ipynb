```python
def call_web(_source, save_html = False):
    """
    Retrieves the HTML content from the specified source URL and optionally saves it to a file.

    Args:
        _source (str): The URL of the web page to retrieve.
        save_html (bool, optional): Indicates whether to save the HTML content to a file. 
                                    Defaults to False.

    Returns:
        bytes: The raw HTML content of the web page.

    """
    req = Request(_source, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36'})
    raw_html = urlopen(req, timeout=10).read()
    
    if save_html == True:
        file = open('../std/my_web.html', 'wb')
        file.write(raw_html)
        file.close()

    return raw_html
```

```python
def parse_url(_raw):
    """
    Parse the raw HTML content and extract the tables.

    Args:
        _raw (bytes): The raw HTML content as a byte string.

    Returns:
        ResultSet: A ResultSet object containing the extracted tables.
    """
    soupdata = BeautifulSoup(_raw, "html.parser")
    
    tables = soupdata.find_all("tr")
    
    #tables = soupdata.find_all('div')
    return tables
```
